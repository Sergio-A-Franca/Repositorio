# Regressão logística

A **Regressão Logística** é uma técnica essencial para problemas de classificação binária, onde precisamos prever a probabilidade de um evento pertencer a uma de duas possíveis classes, como "sim" ou "não", "positivo" ou "negativo". Ao contrário da regressão linear, que lida com a previsão de valores contínuos, a regressão logística é projetada para prever probabilidades associadas a categorias.

### Fundamentos da Regressão Logística

Na regressão logística, nosso objetivo é modelar a **probabilidade** de uma observação pertencer a uma classe específica. Por exemplo, podemos querer prever a probabilidade de um paciente ter uma determinada doença com base em características como idade, peso, e resultados de exames.

#### Função Sigmoide

O coração da regressão logística é a **função sigmoide**. Esta função transforma uma combinação linear de variáveis em uma probabilidade que varia entre 0 e 1. A função sigmoide é matematicamente definida como:

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

Aqui, $ z $ é uma combinação linear das características do modelo, expressa como:

$$
z = w_0 + w_1 x_1 + w_2 x_2 + \dots + w_n x_n
$$

onde:
- $ w_0 $ é o **termo de intercepto**, que é um valor constante,
- $ w_1, w_2, \dots, w_n $ são os **coeficientes** que multiplicam cada característica $ x_1, x_2, \dots, x_n $.

A função sigmoide transforma essa combinação linear em uma probabilidade:

$$
P(y=1|x) = \frac{1}{1 + e^{-(w_0 + w_1 x_1 + \dots + w_n x_n)}}
$$

Essa equação nos dá a probabilidade de uma observação $ x $ pertencer à classe 1.

### Log-Odds e Função Logística

Para entender melhor a regressão logística, é útil compreender o conceito de **log-odds**. A **odds** (razão de chances) é a razão entre a probabilidade de um evento acontecer e a probabilidade de não acontecer. Matematicamente:

$$
\text{odds} = \frac{P(y=1|x)}{1 - P(y=1|x)}
$$

O **log-odds** é simplesmente o logaritmo natural dessa odds:

$$
\text{logit}(P) = \log\left(\frac{P(y=1|x)}{1 - P(y=1|x)}\right)
$$

Na regressão logística, modelamos o log-odds como uma combinação linear das características:

$$
\text{logit}(P) = w_0 + w_1 x_1 + \dots + w_n x_n
$$

Isso significa que, em vez de modelarmos diretamente a probabilidade $ P(y=1|x) $, estamos modelando o logaritmo da odds dessa probabilidade. A vantagem disso é que a relação entre as características e o log-odds é linear, o que facilita a análise e a interpretação dos coeficientes $ w $.

### Estimação dos Coeficientes: Máxima Verossimilhança

Para determinar os valores dos coeficientes $ w $ que melhor ajustam o modelo aos dados, utilizamos um método chamado **Máxima Verossimilhança**. A ideia por trás desse método é encontrar os coeficientes que tornam os dados observados os mais prováveis possíveis, dado o modelo que estamos usando.

A **verossimilhança** é uma função que mede o quão bem o modelo ajusta os dados. Para a regressão logística, a função de verossimilhança $ L $ é:

$$
L(w) = \prod_{i=1}^{m} P(y^{(i)}|x^{(i)};w)
$$

Aqui, $ m $ é o número de observações no conjunto de dados, $ y^{(i)} $ é o rótulo verdadeiro para a $ i $-ésima observação, e $ P(y^{(i)}|x^{(i)};w) $ é a probabilidade prevista pelo modelo para essa observação.

No entanto, como o produto de muitas probabilidades pode se tornar muito pequeno, trabalhamos com o logaritmo da função de verossimilhança, chamado de **log-verossimilhança**:

$$
\log L(w) = \sum_{i=1}^{m} \left[y^{(i)} \log(P(y=1|x^{(i)})) + (1 - y^{(i)}) \log(1 - P(y=1|x^{(i)}))\right]
$$

A maximização dessa função de log-verossimilhança em relação aos coeficientes $ w $ nos dá os valores que melhor ajustam o modelo aos dados.

Na regressão logística, a função de custo mais comum é a **função de perda logarítmica** (também conhecida como **cross-entropy loss**), que é derivada diretamente da log-verossimilhança, mas com um sinal negativo:

$$
J(w) = -\frac{1}{m} \sum_{i=1}^{m} \left[y^{(i)} \log(P(y=1|x^{(i)})) + (1 - y^{(i)}) \log(1 - P(y=1|x^{(i)}))\right]
$$

Aqui, $ J(w) $ é a função de custo que queremos minimizar. Ela mede a diferença entre as previsões do modelo e os rótulos reais, penalizando previsões incorretas. Essa função é simplesmente a log-verossimilhança negativa normalizada pela quantidade de dados.

### Relação entre Maximização da Log-Verossimilhança e Minimização da Função de Custo

Maximizar a log-verossimilhança $ \log L(w) $ é equivalente a **minimizar a função de custo $ J(w) $**. Isso ocorre porque:

- A maximização de uma função positiva (log-verossimilhança) é o mesmo que a minimização de sua negativa (função de custo).
- O sinal negativo na definição da função de custo faz com que o processo de minimização $ J(w) $ seja matematicamente equivalente à maximização de $ \log L(w) $.

### Interpretação dos Coeficientes

Os coeficientes $ w_j $ na regressão logística têm uma interpretação muito interessante. Cada coeficiente representa a mudança no log-odds da probabilidade de $ y=1 $ para uma unidade de aumento na característica $ x_j $. Em outras palavras, $ w_j $ indica como uma pequena mudança em $ x_j $ afeta a probabilidade de pertencimento à classe 1.

Se $ w_j = 0.5 $, por exemplo, isso significa que um aumento de uma unidade em $ x_j $ aumenta o log-odds em 0.5. O impacto direto nas odds pode ser expresso como $ e^{0.5} \approx 1.65 $, ou seja, as chances de $ y=1 $ são multiplicadas por 1.65 para cada unidade adicional de $ x_j $.

### Aplicações e Importância

A regressão logística é amplamente utilizada em diversas áreas. Em medicina, pode ser usada para prever a probabilidade de um paciente ter uma doença com base em sintomas e histórico. Em marketing, pode prever a probabilidade de um cliente realizar uma compra com base em comportamento passado. Seu uso é tão amplo porque fornece uma maneira robusta e interpretável de lidar com problemas de classificação binária.

A regressão logística é um ponto de partida comum para a modelagem de classificações antes de avançar para técnicas mais complexas.

## Um exemplo simples

```python
import matplotlib.pyplot as plt
import numpy as np

COLORMAP = "coolwarm"
```

```python
X = np.array([[2.5, 2.5], [4, -1], [1, -4], [-3, 1.25], [-2, -4], [1, 5]])
y = np.array([1, 1, 1, 0, 0, 0])
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:, 0], X[:, 1], s=100, c=y, cmap=COLORMAP)


plt.xlabel('$x_1$')
plt.ylabel('$x_2$')

plt.show()
```

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(penalty=None)

clf.fit(X, y)
```

```python
clf.coef_
```

```python
clf.intercept_
```

```python
clf.classes_
```

```python
clf.predict([[-2, 2]])
```

```python
clf.predict([[3, 2]])
```

```python
decision_function = clf.decision_function(X)

decision_function
```

```python
X.dot(clf.coef_[0]) + clf.intercept_
```

```python
def funcao_sigmoide(z):
    return 1 / (1 + np.exp(-z))
```

```python
funcao_sigmoide(decision_function)
```

```python
clf.predict_proba(X)
```

```python
clf.predict_proba([[-2, 2]])
```

```python
clf.predict_proba([[3, 2]])
```

```python
odds = clf.predict_proba(X)[:, 1] / (1 - clf.predict_proba(X)[:, 1])

odds
```

```python
clf.predict_proba(X)[:, 1] / clf.predict_proba(X)[:, 0]
```

```python
log_odds = np.log(odds)

log_odds
```

```python
decision_function
```

```python
np.exp(log_odds)
```

```python
from sklearn.inspection import DecisionBoundaryDisplay

fig, ax = plt.subplots()

disp1 = DecisionBoundaryDisplay.from_estimator(
    clf, 
    X, 
    response_method="predict", 
    cmap=COLORMAP, 
    grid_resolution=300, 
    alpha=0.1,
    ax=ax
)

disp2 = DecisionBoundaryDisplay.from_estimator(
    clf, 
    X, 
    response_method="predict", 
    grid_resolution=500, 
    alpha=0.5,
    ax=ax,
    plot_method="contour",
    colors="black",
    linewidths=0.25,
)

ax.scatter(X[:, 0], X[:, 1], c=y, cmap=COLORMAP, s=100)
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")

for i, x in enumerate(X):
    plt.annotate(i, (x[0] + 0.1, x[1] + 0.1))

plt.show()
```

```python
fig, ax = plt.subplots()

disp1 = DecisionBoundaryDisplay.from_estimator(
    clf, 
    X, 
    response_method="decision_function", 
    cmap=COLORMAP, 
    grid_resolution=300, 
    alpha=0.3,
    ax=ax
)

disp2 = DecisionBoundaryDisplay.from_estimator(
    clf, 
    X, 
    response_method="predict", 
    grid_resolution=500, 
    alpha=0.5,
    ax=ax,
    plot_method="contour",
    colors="black",
    linewidths=0.25,
)

ax.scatter(X[:, 0], X[:, 1], c=y, cmap=COLORMAP, s=100)
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")

for i, x in enumerate(X):
    plt.annotate(i, (x[0] + 0.1, x[1] + 0.1))

fig.colorbar(disp1.surface_, ax=ax)

plt.show()
```

```python
fig, ax = plt.subplots()

disp1 = DecisionBoundaryDisplay.from_estimator(
    clf, 
    X, 
    response_method="predict_proba", 
    cmap=COLORMAP, 
    grid_resolution=300, 
    alpha=0.3,
    ax=ax
)

disp2 = DecisionBoundaryDisplay.from_estimator(
    clf, 
    X, 
    response_method="predict", 
    grid_resolution=500, 
    alpha=0.5,
    ax=ax,
    plot_method="contour",
    colors="black",
    linewidths=0.25,
)

ax.scatter(X[:, 0], X[:, 1], c=y, cmap=COLORMAP, s=100)
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")

for i, x in enumerate(X):
    plt.annotate(i, (x[0] + 0.1, x[1] + 0.1))

fig.colorbar(disp1.surface_, ax=ax)

plt.show()
```

```python
def funcao_perda(y_true, y_pred):
    return - np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred)) 
```

```python
y
```

```python
y_pred = clf.predict_proba(X)[:, 1]

y_pred
```

```python
funcao_perda(y, y_pred)
```

```python
from sklearn.metrics import log_loss

log_loss(y, y_pred)
```

```python
clf.score(X, y)
```
