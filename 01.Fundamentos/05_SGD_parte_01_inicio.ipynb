# SGD

O Stochastic Gradient Descent (SGD), ou Gradiente Descendente Estocástico, é uma variação do algoritmo de Gradiente Descendente tradicional que é mais eficiente para conjuntos de dados grandes.

Link Scikit-Learn: https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.SGDClassifier.html

## Significado de Estocástico:

O termo "estocástico" refere-se a processos que têm algum grau de aleatoriedade. No contexto do SGD, "estocástico" refere-se à escolha aleatória de um exemplo de treinamento (ou um pequeno lote de exemplos) para calcular o gradiente e atualizar os parâmetros do modelo em cada etapa, em vez de usar todo o conjunto de dados.

## Diferenças entre SGD e Gradiente Descendente Básico:

1. **Cálculo do Gradiente**:
   - **Gradiente Descendente Básico (GD)**: Calcula o gradiente usando todo o conjunto de dados. Isso é computacionalmente intensivo, especialmente para conjuntos de dados grandes.
   - **Stochastic Gradient Descent (SGD)**: Calcula o gradiente usando apenas um exemplo de treinamento selecionado aleatoriamente (ou um pequeno lote de exemplos) a cada etapa. Isso torna o SGD muito mais rápido.

2. **Convergência**:
   - **GD**: A convergência para o mínimo global é suave e direta porque usa todas as informações do conjunto de dados a cada etapa.
   - **SGD**: A trajetória de convergência pode ser muito mais ruidosa devido à natureza estocástica (aleatória) da atualização dos parâmetros. No entanto, apesar da natureza ruidosa, o SGD muitas vezes convergirá mais rápido, especialmente em grandes conjuntos de dados.

3. **Eficiência Computacional**:
   - **GD**: Pode ser muito lento para grandes conjuntos de dados devido à necessidade de calcular o gradiente em todo o conjunto de dados a cada etapa.
   - **SGD**: É mais eficiente computacionalmente e pode lidar com conjuntos de dados muito grandes.

4. **Variação do Erro**:
   - **GD**: O erro diminui de maneira estável, mas pode ser lento.
   - **SGD**: O erro pode flutuar devido à natureza estocástica das atualizações, mas geralmente alcança uma boa solução mais rapidamente.

O SGD oferece uma maneira eficaz e eficiente de otimizar funções de custo, especialmente em cenários com grandes volumes de dados ou quando os recursos computacionais são limitados.

```python
import matplotlib.pyplot as plt
import numpy as np

COLORMAP = "coolwarm"
```

```python
X = np.array([[2.5, 2.5], [4, -1], [1, -4], [-3, 1.25], [-2, -4], [1, 5]])
y = np.array([1, 1, 1, 0, 0, 0])
```

```python
plt.scatter(X[:, 0], X[:, 1], s=100, c=y, cmap="coolwarm")

plt.xlabel("$x_1$")
plt.ylabel("$x_2$")

plt.show()
```

```python
from sklearn.linear_model import SGDClassifier

clf = SGDClassifier(loss="log_loss", penalty=None, random_state=42)

clf.fit(X, y)
```

```python
clf.coef_
```

```python
clf.intercept_
```

```python
clf.n_iter_
```

```python
clf.decision_function(X)
```

```python
clf.predict(X)
```

```python
clf.predict_proba(X)
```

```python
from sklearn.metrics import log_loss

log_loss(y, clf.predict_proba(X))
```

```python
from sklearn.inspection import DecisionBoundaryDisplay

fig, ax = plt.subplots()

disp1 = DecisionBoundaryDisplay.from_estimator(
    clf, 
    X, 
    response_method="predict", 
    cmap=COLORMAP, 
    grid_resolution=300, 
    alpha=0.1,
    ax=ax
)

disp2 = DecisionBoundaryDisplay.from_estimator(
    clf, 
    X, 
    response_method="predict", 
    grid_resolution=500, 
    alpha=0.5,
    ax=ax,
    plot_method="contour",
    colors="black",
    linewidths=0.25,
)

ax.scatter(X[:, 0], X[:, 1], c=y, cmap=COLORMAP, s=100)
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")

for i, x in enumerate(X):
    plt.annotate(i, (x[0] + 0.1, x[1] + 0.1))

plt.show()
```

```python
fig, ax = plt.subplots()

disp1 = DecisionBoundaryDisplay.from_estimator(
    clf, 
    X, 
    response_method="decision_function", 
    cmap=COLORMAP, 
    grid_resolution=300, 
    alpha=0.3,
    ax=ax
)

disp2 = DecisionBoundaryDisplay.from_estimator(
    clf, 
    X, 
    response_method="predict", 
    grid_resolution=500, 
    alpha=0.5,
    ax=ax,
    plot_method="contour",
    colors="black",
    linewidths=0.25,
)

ax.scatter(X[:, 0], X[:, 1], c=y, cmap=COLORMAP, s=100)
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")

for i, x in enumerate(X):
    plt.annotate(i, (x[0] + 0.1, x[1] + 0.1))

fig.colorbar(disp1.surface_, ax=ax)

plt.show()
```

```python
fig, ax = plt.subplots()

disp1 = DecisionBoundaryDisplay.from_estimator(
    clf, 
    X, 
    response_method="predict_proba", 
    cmap=COLORMAP, 
    grid_resolution=300, 
    alpha=0.3,
    ax=ax
)

disp2 = DecisionBoundaryDisplay.from_estimator(
    clf, 
    X, 
    response_method="predict", 
    grid_resolution=500, 
    alpha=0.5,
    ax=ax,
    plot_method="contour",
    colors="black",
    linewidths=0.25,
)

ax.scatter(X[:, 0], X[:, 1], c=y, cmap=COLORMAP, s=100)
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")

for i, x in enumerate(X):
    plt.annotate(i, (x[0] + 0.1, x[1] + 0.1))

fig.colorbar(disp1.surface_, ax=ax)

plt.show()
```
