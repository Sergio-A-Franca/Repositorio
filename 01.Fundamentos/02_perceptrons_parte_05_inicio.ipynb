# Perceptrons

Perceptrons são a base de muitos modelos de aprendizado de máquina. Eles consistem em um conjunto de entradas, um conjunto de pesos e uma função de ativação. O perceptron recebe um conjunto de entradas, multiplica cada entrada por seu peso e passa a soma total através da função de ativação para obter uma saída ou previsão.



Matematicamente, a operação de um perceptron pode ser descrita da seguinte maneira:

1. **Combinação Linear**:
   A entrada para o perceptron é uma combinação linear dos valores de entrada ($x$) e pesos associados ($w$), junto com um viés ($b$):

$$ z = \sum_{i=1}^n w_i x_i + b $$

onde:
   - $ n $ é o número de entradas,
   - $ w_i $ é o peso associado à $i$-ésima entrada,
   - $ x_i $ é o valor da $i$-ésima entrada,
   - $ b $ é o viés.

2. **Função de Ativação**:

Uma função de ativação é uma função matemática que transforma o valor de saída.

   A saída do perceptron é então obtida aplicando uma função de ativação ($ \phi $) à combinação linear:

$$ y = \phi(z) $$

A função de ativação mais comum usada em um perceptron simples é a função degrau, que pode ser escrita como:

$$ \phi(z) = 
  \begin{cases} 
    1 & \text{se } z \geq 0 \\
    0 & \text{se } z < 0 
  \end{cases}
$$




3. **Regra de Aprendizado**:
   A atualização dos pesos em um perceptron é feita de acordo com a seguinte regra, onde $ \nu $ é a taxa de aprendizado:

$$ w_i^{(n+1)} = w_i^{(n)} + \nu (y - \hat{y}) x_i $$

onde:
   - $ w_i^{(n)} $ é o peso da $i$-ésima entrada no $n$-ésimo passo,
   - $ y $ é o rótulo real,
   - $ \hat{y} $ é a saída prevista do perceptron,
   - $ x_i $ é o valor da $i$-ésima entrada.


## Um exemplo simples

```python
import matplotlib.pyplot as plt
import numpy as np

COLORMAP = "coolwarm"
```

```python
X = np.array([[2.5, 2.5], [4, -1], [1, -4], [-3, 1.25], [-2, -4], [1, 5]])
y = np.array([1, 1, 1, 0, 0, 0])
```

```python
X
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=100, cmap=COLORMAP)

plt.xlabel("$x_1$")
plt.ylabel("$x_2$")

plt.show()
```

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Perceptron.html

```python
from sklearn.linear_model import Perceptron

clf = Perceptron()

clf.fit(X, y)
```

```python
clf.coef_
```

```python
clf.intercept_
```

```python
clf.classes_
```

```python
clf.predict([[-2, 2]])
```

```python
clf.predict([[3, 2]])
```

Agora, podemos usar a função de ativação para obter a saída do perceptron:

$$ \hat{y} = \phi(z) $$

onde $\phi(z)$ é a função de ativação degrau:

$$ 
\phi(z) = 
  \begin{cases} 
    1 & \text{se } z \geq 0 \\
    0 & \text{se } z < 0 
  \end{cases}
$$

```python
pesos = clf.coef_[0]

pesos
```

```python
bias = clf.intercept_

bias
```

```python
X
```

```python
X.dot(pesos) + bias
```

```python
clf.decision_function(X)
```

```python
decision_function = clf.decision_function(X)  # z

decision_function
```

```python
def funcao_degrau(z):
    return np.where(z >= 0, 1, 0)
```

```python
funcao_degrau(decision_function)
```

```python
clf.predict(X)
```

```python
clf.score(X, y)
```

O `score` retornado pelo perceptron é a acurácia do modelo. A acurácia é a proporção de previsões corretas feitas pelo modelo. Podemos calcular a acurácia da seguinte maneira:

$$ \text{Acurácia} = \frac{\text{Número de previsões corretas}}{\text{Número total de previsões}} $$





Há outras métricas de avaliação de modelos que podem ser usadas para avaliar a eficácia de um modelo de aprendizado de máquina. Alguns exemplos incluem precisão, recall, F1-score, matriz de confusão, curva ROC, etc. Veremos essas métricas em detalhes em aulas posteriores.

Na nossa notação, o perceptron pode ser escrito como:

$$ z = w_1 x_1 + w_2 x_2 + b $$

Com os valores de `clf.coef_` e `clf.intercept_` podemos calcular o valor de $z$ para cada ponto de dados:

$$ z = 9.5 \times x_1 - 1.5 \times x_2 - 4.0 $$





Se formos escrever na forma de equação de reta que aprendemos no ensino médio, teremos:

$$ x_2 = -\frac{w_1}{w_2} x_1 - \frac{b}{w_2} $$

```python
pesos, bias
```

```python
inclinacao_reta = - pesos[0] / pesos[1]
intercepto_reta = - bias / pesos[1]

inclinacao_reta, intercepto_reta
```

https://scikit-learn.org/stable/modules/generated/sklearn.inspection.DecisionBoundaryDisplay.html

```python
from sklearn.inspection import DecisionBoundaryDisplay

disp = DecisionBoundaryDisplay.from_estimator(clf, X, response_method="predict", cmap=COLORMAP, grid_resolution=300, alpha=0.5)

disp.ax_.scatter(X[:, 0], X[:, 1], c=y, cmap=COLORMAP, s=100)

disp.ax_.set_xlabel("$x_1$")
disp.ax_.set_ylabel("$x_2$")

for i, x in enumerate(X):
    plt.annotate(i, (x[0] + 0.1, x[1] + 0.1))

plt.show()
```
